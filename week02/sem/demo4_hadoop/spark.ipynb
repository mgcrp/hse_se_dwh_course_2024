```python
from pyspark.sql import SparkSession

namenode = 'hdfs://rc1a-dataproc-m-aw473e4iohpzkb15.mdb.yandexcloud.net'

spark = SparkSession.builder.getOrCreate()

df = spark.read \
    .option('header', 'true') \
    .csv(f'{namenode}/data/mgcrp/breweries.csv')

df.groupby('state') \
    .count() \
    .repartition(1) \
    .write \
    .mode('overwrite') \
    .option('header', 'true') \
    .csv(f'{namenode}/data/mgcrp/breweries_groupby_pySpark.csv')
```
